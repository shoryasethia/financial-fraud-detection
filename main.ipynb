## 1.2 Dealing With Outliers

```python
for cols in df.columns[:-1]:
    sns.boxplot(df['Class'], df[cols])
    plt.xlabel('Class', fontweight = 'bold')
    plt.ylabel(f'{cols}', fontweight = 'bold')
    plt.title(f'{cols} vs Class Boxplot', fontweight = 'bold')
    plt.show()
```

What we can do is, set thresholds for outliers of each variable, by means of a multiple (usually 1.5) of the IQR (Inter-Quartile Range)

```python
# Demo code for dropping outliers, haven't implemented in actual case because many would have to be dropped
    
k = 3 #Threshold, Boxplot has k=1.5
    
df1 = df.copy()
for col in pca_cols :
    # col0, col1 = df_0[col], df_1[col]
    q1, q3 = df.groupby('Class')[col].quantile(0.25)[0], df.groupby('Class')[col].quantile(0.75)[0]
    #q3_0, q3_1 = df.groupby('Class')[col].quantile(0.75)
    iqr = q3-q1 #Interquartile Ranges
    up, low = q3+k*iqr, q1-k*iqr
    
    #Drop all the outliers
    drop_index = df[((df[col]>up) | (df[col]<low)) & (df['Class']==0)].index
    df1.drop(drop_index, inplace = True, errors = 'ignore')
    print(f'Rows Dropped for {col} = {len(drop_index)}\n')
   
m,n = df.shape
print(1-df1.shape[0]/m, df1.shape[0], m)
```

## b) Gaussian Anomaly Detection Algorithm

```python
#To see if columns have different distributions for y=1 or y=0
for col in df.columns[:-1]:
    sns.distplot(df[df['Class']==0].loc[:,col], color='orange', label='Non Fraud', bins = 50)
    sns.distplot(df[df['Class']==1].loc[:,col], color='blue',label='Fraud', bins = 50)
    plt.legend()
    plt.title(f'Fraud vs Non-Fraud Distribution of {col}', fontweight = 'bold')
    plt.show()
```

```python
drop_columns = ['V2','V5','V6','V7','V8','V13','V15','V20','V21','V22','V23','V24','V25','V26','V27','V28','Amount']
#No Fraud
df_nf = df2[df2['Class']==0].drop(drop_columns, axis = 1).copy()
#Fraud
df_f = df2[df2['Class']==1].drop(drop_columns, axis = 1).copy()

#NON-FRAUD CASES
#Separating Test Data
df_t, df_nf_test = train_test_split(df_nf, test_size = 0.2, random_state = 0)
#Separating Training and CV Data
df_train, df_nf_cv = train_test_split(df_t, test_size = 0.25, random_state = 0)

#FRAUD CASES
#Separating Cross Validation and Test Data
df_f_cv, df_f_test = train_test_split(df_f, test_size = 0.5)

#Finally Creating CrossVal and Test df
df_cv = pd.concat([df_nf_cv,df_f_cv]).sample(frac=1)
df_test = pd.concat([df_nf_test, df_f_test]).sample(frac=1)

#Separating Feature Matrix & Ground Truths
X_train = df_train.drop('Class', axis = 1)
X_cv, y_cv = df_cv.drop('Class', axis = 1), df_cv['Class']
X_test, y_test = df_test.drop('Class', axis = 1), df_test['Class']
```

```python
from scipy.stats import multivariate_normal

#Defining the GAD Algorithm
def normal_params(X):
    mu = np.mean(X, axis = 0)
    Sigma = np.cov(X.T) #Covariance Matrix
    return mu, Sigma

mu_train, Sigma_train = normal_params(X_train)

def anomalyDetection(X):
    prob = multivariate_normal.pdf(X, mu_train, Sigma_train, allow_singular = True)
    return prob**(1/100) #Since prob values are very very small

#Training Data must be non anomalous
p_train = anomalyDetection(X_train)
```

```python
#Cross Validation to determine the value of epsilon
def crossVal(prob,y):
    best_epsilon = 0
    best_precision, best_recall = 0,0
    best_f1 = 0
    
    epsilons = sorted(np.unique([np.round(i,3) for i in prob]))
    precisions=[]
    recalls=[]
    
    for epsilon in epsilons:
        predictions = (prob < epsilon)
        f = f1_score(y, predictions)
        precision = precision_score(y, predictions)
        recall = recall_score(y, predictions)
        
        precisions.append(precision)
        recalls.append(recall)
        
        if f>best_f1:
            best_f1 = f
            best_precision = precision
            best_recall = recall
            best_epsilon = epsilon
    
    #Plotting the Precision-Recall Curve
    plt.plot(epsilons,precisions,label='Precision')
    plt.plot(epsilons,recalls,label='Recall')
    plt.xlabel('Epsilon', fontweight = 'bold')
    plt.title('Precision-Recall Curve')
    plt.legend()
    plt.show()

    print ('Best F1 Score %f' %best_f1)
    print ('Best Precision Score %f' %best_precision)
    print ('Best Recall Score %f' %best_recall)
    print ('Best Epsilon', best_epsilon) 
    
    return best_epsilon
```

```python
prob = anomalyDetection(X_cv)
crossVal(prob, y_cv)
```

```python
from sklearn.metrics import confusion_matrix

#Testing the Model on Test Dataset
epsilon = 0.222
prob_test = anomalyDetection(X_test)
y_test_pred = prob_test<epsilon

f = f1_score(y_test, y_test_pred)
p = precision_score(y_test, y_test_pred)
r = recall_score(y_test, y_test_pred)

print('F1 Score = ',f)
print('Precision = ',p)
print('Recall = ',r)

sns.heatmap(confusion_matrix(y_test, y_test_pred), annot = True, fmt='d')
plt.ylabel('Ground Truth', fontweight = 'bold')
plt.xlabel('Predictions', fontweight = 'bold')
plt.title('Confusion Matrix', fontweight = 'bold')
plt.show()
```

```python
X_test = data_test.drop('Class', axis = 1)
y_test = data_test['Class']

#Separating Fraud and Non Fraud ex. in Train Data
data_train_nf = data_train[data_train['Class']==0]
data_train_f = data_train[data_train['Class']==1]

#Total number of Models to be trained
m_train_f = data_train_f.shape[0]
m_train_nf = data_train_nf.shape[0]
# n_models = int(m_train_nf/m_train_f)

#Creating Training Data for each Model
list_train_data = []
for i in range(0,m_train_nf, 19*m_train_f):
    t = data_train_f.append(data_train_nf[i:i+m_train_f])
    
    #Randomizing the arrangement
    train_sample = t.sample(frac = 1)
    del t
    #Appending it to our data list
    list_train_data.append(train_sample)
    
n_models = len(list_train_data)
```

```python
from sklearn.neighbors import KNeighborsClassifier
#Training all the DT Models
model_list = []
for i in range(n_models):
    #Defining Train Data
    X_train = list_train_data[i].drop('Class',axis = 1)
    y_train = list_train_data[i]['Class']
    
    #Initializing, Fitting and Storing Model
    dt_model = KNeighborsClassifier(n_neighbors = 41, n_jobs = -1)
    dt_model.fit(X_train, y_train)
    model_list.append(dt_model)
```

```python
#def predict(X_test):
predictions = []

#Predicting on the Test Data
for i in range(n_models):
    y_test_pred = model_list[i].predict(X_test)
    predictions.append(y_test_pred)

test_pred_df = pd.DataFrame(np.array(predictions).T)
pred = test_pred_df.mode(axis = 1)
```
